```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('data.csv')
```

```python
df = df.drop(columns = 'Unnamed: 32')#removing the null feature
```

```python
df = pd.get_dummies(data = df, drop_first = True) #one hot encoding
```

```python
df["diagnosis_M"] = df["diagnosis_M"].astype(int) #one hot encoding as 0 or 1
```

```python
x=df.iloc[:,1:-1] #Features
```

```python
y=df.iloc[:,-1] #classifications
```

```python
from sklearn.model_selection import train_test_split 
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2, random_state=0)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
sc = StandardScaler()
```

```python
x_train = sc.fit_transform(x_train)
```

```python
x_test=sc.transform(x_test)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
classifier = LogisticRegression(random_state=0,C=1, penalty = 'l2', solver = 'lbfgs')
```

```python
classifier.fit(x_train,y_train)
```

```python
y_pred = classifier.predict(x_test)
```

```python
from sklearn.metrics import precision_score
from sklearn.metrics import accuracy_score
from sklearn.metrics import f1_score
from sklearn.metrics import recall_score
from sklearn.metrics import confusion_matrix
```

```python
acc = accuracy_score(y_test,y_pred)
f1 = f1_score(y_test, y_pred)
rec = recall_score(y_test, y_pred)
pre = precision_score(y_test, y_pred)
```

```python
final_result = pd.DataFrame([['logistic regression',acc,f1,rec,pre]],columns = ['model','accuracy','f1','recall','precision'])
```

```python
from sklearn.model_selection import cross_val_score
```

```python
accu=cross_val_score(estimator = classifier, X=x_train, y = y_train)
```

```python
print(f"accuracy is: {accu.mean()*100}")
```
